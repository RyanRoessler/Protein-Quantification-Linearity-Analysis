```python
from matplotlib.ticker import FormatStrFormatter, StrMethodFormatter

# Sample 1-1

# VISUALIZING RANGEES

# Initializing x and y data (Sample 1-1)
x = df_1_1.Concentration
y = df_1_1.oncoprotein_area
Max = 1.28

# Initializing figures/subplots
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(30, 9))
fig.tight_layout(pad=9)

# PLOT 1, Range 1
ax1.plot(x, y, color='green', linewidth=5, marker='o', markersize=10, markerfacecolor='black', label='Range 1')
ax1.legend(loc='upper left', fontsize='xx-large')
ax1.set_title(r'Sample 1-1: oncoprotein Area', fontweight="bold", fontsize=30)
ax1.set_xlabel('Concentration (ng/uL)', fontsize=30, labelpad=10) # labelpad sets distance of axis label from tick labels
ax1.set_ylabel('oncoprotein Area', fontsize=30, labelpad=10)
# Scientific Notation (y-axis)
ax1.ticklabel_format(axis='y', style='sci', scilimits=(6,6))
# Font size for "1e6"
ax1.yaxis.offsetText.set_fontsize(24)
# Sets font size of ticks
ax1.tick_params(labelsize=25)
# Labels only the 4 concentrations used on the x-axis
ax1.set_xticks(x)
# Grid
ax1.grid(linestyle='--', linewidth=2)
# Range visualization (x-axis)
ax1.axvspan(0.1, Max, color='yellow', alpha=0.3)

# PLOT 2, Range 2
ax2.plot(x, y, color='green', linewidth=5, marker='o', markersize=10, markerfacecolor='black', label='Range 2')
ax2.legend(loc='upper left', fontsize='xx-large')
ax2.set_title(r'Sample 1-1: oncoprotein Area', fontweight="bold", fontsize=30)
ax2.set_xlabel('Concentration (ng/uL)', fontsize=30, labelpad=10) # Labelpad sets distance of axis label from tick labels
ax2.set_ylabel('oncoprotein Area', fontsize=30, labelpad=10)
# Scientific Notation (y-axis)
ax2.ticklabel_format(axis='y', style='sci', scilimits=(6,6))
# Font size for "1e6"
ax2.yaxis.offsetText.set_fontsize(24)
# Sets font size of ticks
ax2.tick_params(labelsize=25)
# Labels only the 4 concentrations used on the x-axis
ax2.set_xticks(x)
# Grid
ax2.grid(linestyle='--', linewidth=2)
# Range visualization (x-axis)
ax2.axvspan(0.1, 1.0, color='yellow', alpha=0.3)

#Plot 3, Range 3
ax3.plot(x, y, color='green', linewidth=5, marker='o', markersize=10, markerfacecolor='black', label='Range 3')
ax3.legend(loc='upper left', fontsize='xx-large')
ax3.set_title(r'Sample 1-1: oncoprotein Area', fontweight="bold", fontsize=30)
ax3.set_xlabel('Concentration (ng/uL)', fontsize=30, labelpad=10) # labelpad sets distance of axis label from tick labels
ax3.set_ylabel('oncoprotein Area', fontsize=30, labelpad=10)
# Scientific Notation (y-axis)
ax3.ticklabel_format(axis='y', style='sci', scilimits=(6,6))
# Font size for "1e6"
ax3.yaxis.offsetText.set_fontsize(24)
# Sets font size of ticks
ax3.tick_params(labelsize=25)
# Labels only the 4 concentrations used on the x-axis
ax3.set_xticks(x)
# Grid
ax3.grid(linestyle='--', linewidth=2)
# Range visualization (x-axis)
ax3.axvspan(0.5, Max, color='yellow', alpha=0.3)
##########################################################################################


# R_squared PER Range

import statsmodels.api as sm
import statsmodels.formula.api as smf
import seaborn as sns
from sklearn.linear_model import LinearRegression
from scipy import stats

x1 = np.array(df_1_1.Concentration).reshape((-1, 1))
y1 = np.array(df_1_1.oncoprotein_area)
# Max Concentration
print(f"x1 (0.1-{Max}): {x1}")
print(f"y1 (0.1-{Max}): {y1}")

# # Linear Regression
# model = LinearRegression().fit(x, y)
# Obtain R_squared using score()
# r_sq = model.score(x, y)
# print(f"coefficient of determination: {r_sq}")
# # Slope
# print(f"slope: {model.coef_}")
# # y-intercept
# print(f"intercept: {model.intercept_}")


# OBTAINING R_squared FOR VARIOUS RANGES OF DATA TO FIND 'MOST LINEAR' RANGE

# (Range 1) 0.1 - Max
model = LinearRegression().fit(x1, y1)
# R_squared (Range 1)
r_sq1 = model.score(x1, y1)
print(f"(Range 1) 0.1 - {Max}: R = {r_sq1}\n")

# (Range 2) 0.1 - 1
x2 = np.array(df_1_1.iloc[1:,:].Concentration).reshape((-1, 1))
y2 = np.array(df_1_1.iloc[1:,:].oncoprotein_area)
print(f"x2 (0.1-1): {x2}")
print(f"y2 (0.1-1): {y2}")
model = LinearRegression().fit(x2, y2)
# R_squared (Range 2)
r_sq2 = model.score(x2, y2)
print(f"(Range 2) 0.1 - 1: R = {r_sq2}\n")

# (Range 3) 0.5 - Max
x3 = np.array(df_1_1.iloc[:3,:].Concentration).reshape((-1, 1))
y3 = np.array(df_1_1.iloc[:3,:].oncoprotein_area)
print(f"x3 (0.5-{Max}): {x3}")
print(f"y3 (0.5-{Max}): {y3}")
model = LinearRegression().fit(x3, y3)
# R_squared (Range 3)
r_sq3 = model.score(x3, y3)
print(f"(Range 3) 0.5 - {Max}: R = {r_sq3}\n")

# Creating Dictionary of Ranges:R_squared
dict_R = {
    f'(Range 1) 0.1 - {Max}':r_sq1,
    f'(Range 2) 0.1 - 1':r_sq2,
    f'(Range 3) 0.5 - {Max}':r_sq3
}

# Finding/Printing Range with Max R_squared value
max_key = max(dict_R, key=dict_R.get)
max_key_value = dict_R.get(max_key)
print("Range with highest linearity:")
print(f"{max_key}, {max_key_value}")
```
